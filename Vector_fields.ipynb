```python
import ipywidgets as pw
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
pi = np.pi
```

```python
xlim      = [-10,10]
ylim      = [-10,10]
step_size = .1
x, y = np.mgrid[xlim[0]:xlim[1]:step_size,ylim[0]:ylim[1]:step_size]
```

```python
u = np.linspace(-pi, pi, 100)
pw.interact(lambda v:[ plt.quiver(x, y, x, y*np.exp(1j*v)), plt.xlim(*xlim), plt.ylim(*ylim), plt.gcf().set_size_inches(10,5)], v=(-pi, pi,.01 ) )
```

```python
pw.interact(lambda u, v : [plt.quiver(0, 0, np.cos(u), np.sin(u)*np.exp(1j*v), scale= 3),
                           plt.text(-.8,-.8, f'[{np.cos(u):.2f},\n{np.sin(u)*np.exp(1j*v):.2f}]', ), 
                           plt.gca().set(xlim=(-1,1), ylim =(-1,1))], 
                            u = (-np.pi,np.pi,.01),
                            v = (-np.pi,np.pi,.01))
```

```python
fig = plt.figure(figsize= (8,5))
ax  = fig.add_subplot()
x, y = np.mgrid[-5:5.05, -5:5:.05]
u = np.cos(x)
v = np.sin(y)
ax.quiver(x, y, u, v)
```
